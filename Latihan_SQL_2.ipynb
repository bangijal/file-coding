## One to many relationship

```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

### 1. Report the account representative for each customer.

```python
run("""
SELECT 
    customerNumber, 
    customerName, 
    salesRepEmployeeNumber    
FROM
    customers t1
        INNER JOIN
    employees t2 ON t1.salesRepEmployeeNumber = t2.employeeNumber
""")
```

### 2. Report total payments for Atelier graphique.

```python
run("""
SELECT 
    customerName,
    SUM(amount) AS Total_payments
FROM
    customers t1
    INNER JOIN
    payments t2 
        ON t1.customerNumber = t2.customerNumber
WHERE
    customerName='Atelier graphique'
""")
```

### 4. Report the products that have not been sold.

```python
run("""
SELECT 
    productName,
    quantityOrdered
FROM
    orderdetails tb1
    RIGHT JOIN
    products tb2 
        ON tb1.productCode = tb2.productCode
WHERE 
    tb1.orderNumber IS NULL
""")
```

### 8. Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.

## Many to many relationship

### 1. List products sold by order date.

```python
run("""
SELECT 
    (productName) AS Product_Name,    
    orderDate
FROM
    ((orderdetails t1
    INNER JOIN
    orders t2 
        ON t1.orderNumber = t2.orderNumber)
    INNER JOIN
    products t3 
        ON t1.productCode = t3.productCode)  
ORDER BY orderDate 
""")
```

### 2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

```python
run("""
SELECT 
    productName,
    orderDate
FROM
    ((orderdetails t1
    INNER JOIN
    orders t2 
        ON t1.orderNumber = t2.orderNumber)
    INNER JOIN
    products t3 
        ON t1.productCode = t3.productCode)  
WHERE 
    productName='1940 Ford Pickup Truck'
ORDER BY orderDate ASC 
""")
```

### 3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?

```python
run("""
SELECT 
    customerName,
    orderNumber,
    amount
FROM
    ((customers t1
    INNER JOIN
    orders t2 
        ON t1.customerNumber = t2.customerNumber)
    INNER JOIN
    payments t3 
        ON t1.customerNumber = t3.customerNumber)  
 WHERE amount>25000
""")
```

### 4. Are there any products that appear on all orders?

```python
run("""
SELECT 
    productName,
    COUNT(t1.orderNumber) AS number_of_order
FROM
    ((orderdetails t1
    INNER JOIN
    products t2 
        ON t1.productCode = t2.productCode)
    INNER JOIN
    orders t3 
        ON t1.orderNumber = t3.orderNumber)  
GROUP BY productName      
ORDER BY number_of_order
""")
```

### 5. List the names of products sold at less than 80% of the MSRP.

```python
run("""
SELECT 
    productName,
    priceEach,
    MSRP
    
FROM
    products t1
    INNER JOIN
    orderdetails t2 
        ON t1.productCode = t2.productCode
WHERE
    priceEach<(0.8*MSRP)
ORDER BY priceEach
""")
```

### 6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)

```python
run("""
SELECT 
    orderNUmber,
    productName,    
    buyPrice,
    priceEach
FROM
    products t1
    INNER JOIN
    orderdetails t2 
        ON t1.productCode = t2.productCode
WHERE
    priceEach>(2*buyPrice)
ORDER BY priceEach
""")
```

### 7. List the products ordered on a Monday.

```python
run("""
SELECT 
    productName,
    DAYNAME(orderDate) AS Day,
    orderDate
FROM
    ((orderdetails t1
    INNER JOIN
    products t2 
        ON t1.productCode = t2.productCode)
    INNER JOIN
    orders t3 
        ON t1.orderNumber = t3.orderNumber)  
WHERE 
    DAYNAME(orderDate)='Monday'
""")
```
